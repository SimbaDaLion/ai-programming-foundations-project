```python
# Load Iris dataset
df = pd.read_csv('iris.csv')

# Display first 5 rows
df.head()
```

```python
def check_missing_values(df):
    """
    Check for missing values in the dataset.
    
    Parameters:
    df (pandas.DataFrame): The dataset to check
    
    Returns:
    pandas.DataFrame: Summary of missing values per column
    """
    missing = df.isnull().sum()
    print("Missing Values Report:")
    print(missing)
    print(f"\nTotal missing values: {missing.sum()}")
    return missing

# Run the function
check_missing_values(df)
```

```python
def check_data_types(df):
    """
    Display data types and basic info about the dataset.
    
    Parameters:
    df (pandas.DataFrame): The dataset to check
    
    Returns:
    None: Prints data types and column info
    """
    print("Data Types:")
    print(df.dtypes)
    print("\n" + "="*40)
    print("Dataset Info:")
    df.info()

# Run the function
check_data_types(df)
```

```python
def summary_statistics(df):
    """
    Generate summary statistics and group by species.
    
    Parameters:
    df (pandas.DataFrame): The dataset to analyze
    
    Returns:
    None: Prints summary statistics
    """
    print("Summary Statistics:")
    print(df.describe())
    print("\n" + "="*50)
    print("\nGroup by Species - Average Measurements:")
    print(df.groupby('species').mean().round(2))

# Run the function
summary_statistics(df)
```

```python
# Visualization 1: Box plot of petal length by species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal Length Distribution by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.show()
```

```python
# Visualization 2: Scatter plot colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, s=100)
plt.title('Sepal Length vs Sepal Width by Species', fontsize=14)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.legend(title='Species')
plt.show()
```

```python
# Visualization 3: Pairplot showing relationships between all features
sns.pairplot(df, hue='species', height=2.5)
plt.suptitle('Pairplot of Iris Features by Species', y=1.02, fontsize=14)
plt.show()
```

## Summary and Interpretation

### What I Learned
The Iris dataset demonstrates clear morphological differences between three species of iris flowers. The data contains 150 samples with 4 measurements: sepal length, sepal width, petal length, and petal width.

### Interesting Patterns
1. **Petal measurements are the best discriminators**: The box plot (Figure 1) shows petal length clearly separates all three species, with no overlap between setosa and the other two species.
2. **Setosa is distinctly different**: The scatter plot (Figure 2) shows setosa (blue) clusters separately from versicolor (orange) and virginica (green) in sepal dimensions.
3. **Strong correlations exist**: The pairplot (Figure 3) reveals strong positive correlations between petal length and petal width across all species.

### Assumptions and Limitations
- This dataset is small (150 rows) and may not represent all iris varieties
- Measurements were taken under controlled conditions, which may not reflect natural variation
- The dataset is perfectly clean with no missing values, which is not typical of real-world data

### Surprising Findings
The sepal width shows considerable overlap between species, making it the least useful feature for classification. In contrast, petal measurements provide almost perfect separation.
